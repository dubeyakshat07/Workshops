Dealing with imbalanced data using PyCaret

Building PyCaret model without fixing the imbalance problem

```python
exp = setup(data=telecom, target="Churn")
```

```python
compare_models()
```

Treating the dataset imablance problem using PyCaret

```python
exp = setup(data=telecom, target="Churn", fix_imbalance = True)
```

```python
compare_models()
```

```python
lightgbm=create_model("lightgbm")
```

```python
tuned_lgbm = tune_model(lightgbm)
```

Steps to perform ML using PyCaret
1. Setup the experiment
2. Compare different models
3. Build the best model according to your best metric
4. Perform hyperparameter tuning

1. exp = setup(data=dataframe_variable, target="target_column)
2. compare_models()
3. model = create_model("model abbreviation")
4. tuned_model = tune_model(model)
